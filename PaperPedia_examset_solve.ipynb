# 3)Age

```python
#We consider here the age_group between
#Child==0-15
#Teenage==15-30
#Adult==30-50
#Oldage==50+ above
```

```python
Age_data=[]
for x in data["Age"]:
    if 0<x<=15:
        Age_data.append("Child")
    elif 15<x<=30:
        Age_data.append("Teenage")
    elif 30<x<=50:
        Age_data.append("Adult")
    else:
        Age_data.append("Oldage")
```

```python
data["Age_data"]=Age_data
```
